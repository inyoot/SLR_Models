## Introduction of Sign Language Recognition Research

### The Dynamism and Complexity of Sign Language

The multimodal communication system of sign language is essentially different from spoken or written languages, as its meaning is inherently rooted in physical, dynamic action. Machine learning models are unable to depend on a single, isolated data stream, such as audio or text, in order to accurately capture and interpret these languages. Rather, they are required to synthesize intricate, concurrent inputs. This necessitates the surveillance of nuanced facial landmarks and broader body posture signals, in addition to high-fidelity hand pose-tracking to map the precise shape and movement of the digits. Every one of these components possesses significant semantic significance; a minor adjustment to the trajectory of a hand or the position of an eyebrow can completely transform the meaning of a sentence. Consequently, the development of effective recognition systems necessitates the implementation of highly advanced monitoring technologies that are capable of interpreting this complex physical vocabulary in real time.

### Survey of the State of The Art of Sign Language Recognition

Due to their capacity to analyze sequential and multimodal data, including multiple-camera video sequences, Hidden Markov Models (HMMs) have gained significant popularity in the field of Sign Language Recognition (SLR). Input devices, feature extraction methods, and a recognizer comprise the standard architecture of an SLR system. The HMM recognizer classifies input, which can range from video files and RGB cameras to depth sensors and electronic gloves, after feature extraction using computer vision or dimension reduction techniques.

The evolution of HMMs in SLR is divided into five chronological episodes, with the first episode being Classical HMMs, which were largely reliant on early computer vision techniques. RGB cameras and color-coded mittens were employed by early systems to monitor hand movements for datasets such as Filipino Sign Language. Depth-sensing technologies, such as the Microsoft Kinect and stereo cameras, were introduced in the second episode, Extended HMMs. Weighted Hidden Markov Models (WHMMs) were created by researchers to resolve the differences between various signers. These models assign varying weights to sign samples based on training errors, thereby enhancing the accuracy of recognition.

The subsequent episodes, beginning with HMMs and Machine Learning, compared and combined HMMs with neural networks and Support Vector Machines, thereby expanding the capabilities of HMMs through advanced computational techniques. The HMMs and Sensor Fusion episode followed, which enhanced 3D feature extraction by integrating data from multiple sensors. For example, the episode involved the integration of Leap Motion devices with Kinect cameras using early and late data fusion techniques. Lastly, the Automatic HMMs and Big Data episode concentrated on the management of large, intricate datasets by enabling the system to independently determine the optimal number of HMM states for various signs and integrating transition models to address the beginning and end of signs in continuous sign language.

HMMs have significantly enhanced SLR over the past decade by incorporating signer independence, sensor fusion, and inherent model selection. In the future, the survey emphasizes a substantial correlation between HMMs and Recurrent Neural Networks (RNNs), indicating that both are essential for the modeling of sequences of random observations. The utilization of modern hardware platforms that include smartphone sensors (gyroscopes, LiDAR) in conjunction with cloud computing and frameworks such as MediaPipe is emphasized in future research directions to create more accessible, real-time sign language translation systems.

```python
# ============================================================
# SAFE PARQUET READER
# ============================================================
def safe_read_parquet(filepath):
    try:
        table = pq.read_table(str(filepath))
        return table.to_pandas()
    except Exception as e:
        return None


# ============================================================
# FEATURE EXTRACTION — RAW COORDS (hands + pose only, X,Y only)
# ============================================================
def extract_raw_coords(df):
    """
    Returns (T, N_RAW) float32 array with raw x,y for
    left_hand(21) + right_hand(21) + pose(33) landmarks.
    Frames are densely remapped from whatever range they use.
    N_RAW = (N_HAND * 2 + N_POSE) * 2 = 150 (since z is excluded)
    """
    if df is None or len(df) == 0:
        return None

    df = df.dropna(subset=['frame']).copy()
    if len(df) == 0:
        return None

    unique_frames = np.sort(df['frame'].unique())
    n_frames = len(unique_frames)
    if n_frames < 4:
        return None

    frame_map = {int(f): i for i, f in enumerate(unique_frames)}
    df['fi'] = df['frame'].map(frame_map).astype(int)
    df['x']  = pd.to_numeric(df['x'], errors='coerce').fillna(0.0).astype(np.float32)
    df['y']  = pd.to_numeric(df['y'], errors='coerce').fillna(0.0).astype(np.float32)
    # Z-coordinates are explicitly excluded
    # df['z']  = pd.to_numeric(df['z'], errors='coerce').fillna(0.0).astype(np.float32)

    # N_RAW is now 150 (75 landmarks * 2 dimensions)
    raw = np.zeros((n_frames, N_RAW), dtype=np.float32)

    # Number of dimensions per landmark (x, y)
    NEW_DIM_PER_LANDMARK = 2

    def fill(mask, offset, max_lm):
        sub = df.loc[mask & (df['landmark_index'] < max_lm)]
        if len(sub) == 0:
            return
        fi = sub['fi'].values.astype(int)
        li = sub['landmark_index'].values.astype(int)
        # Adjust column calculation for only 2 dimensions (x, y)
        bi = offset + li * NEW_DIM_PER_LANDMARK
        # Check bounds: N_RAW refers to total x,y features
        ok = (fi < n_frames) & (bi + 1 < N_RAW) # +1 for y coord
        raw[fi[ok], bi[ok]]     = sub['x'].values[ok]
        raw[fi[ok], bi[ok] + 1] = sub['y'].values[ok]
        # No 'z' coordinate to fill

    # Adjust offsets based on 2 dimensions per landmark
    fill(df['type'] == 'left_hand',  0,                                  N_HAND)
    fill(df['type'] == 'right_hand', N_HAND * NEW_DIM_PER_LANDMARK,      N_HAND)
    fill(df['type'] == 'pose',       2 * N_HAND * NEW_DIM_PER_LANDMARK,  N_POSE)

    # Must have at least some hand signal (adjusted for 2 dimensions)
    hand_region = raw[:, :N_HAND * 2 * NEW_DIM_PER_LANDMARK]
    if np.count_nonzero(hand_region.sum(axis=1)) < 3:
        return None

    return raw   # (T, 150)
```

```python
# ============================================================
# FEATURE ENGINEERING — the critical fix
# ============================================================
def engineer_features(raw):
    """
    Transform raw (T, N_RAW_2D) coords into meaningful features:

    1. RELATIVE coords: subtract wrist position so features are
       translation-invariant (person can sit anywhere)
    2. NORMALISED scale: divide by hand span so size-invariant
    3. VELOCITY: frame-to-frame differences (motion matters!)
    4. ACCELERATION: second-order differences
    5. Select only most informative landmarks

    Output: (T, feat_dim) — much more discriminative
    """
    T = raw.shape[0]

    # --- Z-coordinates are now already filtered by extract_raw_coords ---
    # raw is expected to be (T, (N_HAND*2 + N_POSE)*2) = (T, 150)
    NEW_DIM_PER_LANDMARK = 2 # only x, y

    # Direct assignment, as raw already contains only X,Y coordinates
    raw_2d_flat  = raw

    # ── Indices (adjusted for 2D) ──────────────────────────────
    # Left hand: landmarks 0..20 → cols 0.. (N_HAND * 2 - 1)
    # Right hand: landmarks 0..20 → cols (N_HAND * 2) .. (N_HAND * 4 - 1)
    # Pose: landmarks 0..32 → cols (N_HAND * 4) .. (N_ALL_LANDMARKS * 2 - 1)

    LH_START = 0
    RH_START = N_HAND * NEW_DIM_PER_LANDMARK
    PS_START = 2 * N_HAND * NEW_DIM_PER_LANDMARK

    # Wrist = landmark 0 of each hand (now only x, y)
    lh_wrist_x = raw_2d_flat[:, LH_START + 0]
    lh_wrist_y = raw_2d_flat[:, LH_START + 1]
    rh_wrist_x = raw_2d_flat[:, RH_START + 0]
    rh_wrist_y = raw_2d_flat[:, RH_START + 1]

    # ── 1. Relative hand coords (subtract wrist) ─────────────
    lh_rel = raw_2d_flat[:, LH_START:LH_START + N_HAND * NEW_DIM_PER_LANDMARK].copy()
    rh_rel = raw_2d_flat[:, RH_START:RH_START + N_HAND * NEW_DIM_PER_LANDMARK].copy()

    for i in range(N_HAND):
        lh_rel[:, i*NEW_DIM_PER_LANDMARK]   -= lh_wrist_x
        lh_rel[:, i*NEW_DIM_PER_LANDMARK+1] -= lh_wrist_y
        rh_rel[:, i*NEW_DIM_PER_LANDMARK]   -= rh_wrist_x
        rh_rel[:, i*NEW_DIM_PER_LANDMARK+1] -= rh_wrist_y

    # ── 2. Scale normalisation (by distance middle finger tip to wrist, now 2D) ──
    # LH middle finger tip = landmark 12
    lh_tip = np.sqrt(
        lh_rel[:, 12*NEW_DIM_PER_LANDMARK]**2 + lh_rel[:, 12*NEW_DIM_PER_LANDMARK+1]**2
    ).clip(min=1e-6)
    rh_tip = np.sqrt(
        rh_rel[:, 12*NEW_DIM_PER_LANDMARK]**2 + rh_rel[:, 12*NEW_DIM_PER_LANDMARK+1]**2
    ).clip(min=1e-6)

    lh_rel /= lh_tip[:, None]
    rh_rel /= rh_tip[:, None]

    # ── 3. Pose: use only upper body (landmarks 0-16 = head+shoulders+arms, now 2D) ──
    # N_ALL_LANDMARKS is 75 (21+21+33). 17 landmarks from pose are selected
    N_ALL_LANDMARKS = N_HAND * 2 + N_POSE
    pose_sel_start_idx = 2 * N_HAND * NEW_DIM_PER_LANDMARK # Start of pose in raw_2d_flat
    pose_sel = raw_2d_flat[:, pose_sel_start_idx : pose_sel_start_idx + 17*NEW_DIM_PER_LANDMARK]  # 17 landmarks * 2 = 34

    # Relative to nose (landmark 0 of pose)
    nose_x = pose_sel[:, 0]
    nose_y = pose_sel[:, 1]
    for i in range(17):
        pose_sel[:, i*NEW_DIM_PER_LANDMARK]   -= nose_x
        pose_sel[:, i*NEW_DIM_PER_LANDMARK+1] -= nose_y

    # ── 4. Concatenate spatial features (adjusted for 2D) ─────────────────────
    # lh_rel: (21*2) = 42, rh_rel: (21*2) = 42, pose_sel: (17*2) = 34 → 42+42+34 = 118
    spatial = np.concatenate([lh_rel, rh_rel, pose_sel], axis=1)  # (T, 118)

    # ── 5. Velocity (frame differences) ─────────────────────
    # Pad first row so shape stays (T, 118)
    velocity = np.zeros_like(spatial)
    velocity[1:] = spatial[1:] - spatial[:-1]

    # ── 6. Acceleration ──────────────────────────────────────
    accel = np.zeros_like(spatial)
    accel[2:] = velocity[2:] - velocity[1:-1]

    # ── 7. Absolute wrist positions (where hands are in frame, now 2D) ──
    wrist_pos = np.stack([
        lh_wrist_x, lh_wrist_y,
        rh_wrist_x, rh_wrist_y
    ], axis=1)   # (T, 4)

    # Velocity of wrist
    wrist_vel = np.zeros_like(wrist_pos)
    wrist_vel[1:] = wrist_pos[1:] - wrist_pos[:-1]

    # ── 8. Combine all (adjusted for 2D) ───────────────────────────
    # spatial:118 + velocity:118 + accel:118 + wrist:4 + wrist_vel:4 = 354 + 8 = 362
    features = np.concatenate([
        spatial,      # shape (T, 118)
        velocity,     # shape (T, 118)
        accel,        # shape (T, 118)
        wrist_pos,    # shape (T, 4)
        wrist_vel     # shape (T, 4)
    ], axis=1)        # → (T, 362)

    return features.astype(np.float32)


def resample_seq(seq, target_len):
    """
    Temporally resample to fixed length.
    """
    T = seq.shape[0]
    if T == target_len:
        return seq
    idx = np.round(np.linspace(0, T - 1, target_len)).astype(int)
    return seq[idx]


def per_sequence_normalize(seq):
    """
    Normalize each sequence independently.
    This is critical — global normalization loses relative info.
    Use robust normalization: subtract median, divide by IQR.
    """
    # Only normalize non-zero features
    flat = seq.reshape(-1)
    nonzero = flat[flat != 0]
    if len(nonzero) == 0:
        return seq
    median = np.median(nonzero)
    q75, q25 = np.percentile(nonzero, [75, 25])
    iqr = q75 - q25
    if iqr < 1e-8:
        iqr = nonzero.std() + 1e-8
    return ((seq - median) / iqr).clip(-10, 10).astype(np.float32)
```

```python
# ============================================================
# DATA AUGMENTATION
# ============================================================
def augment_sequence(seq, training=True):
    """
    Apply random augmentations to prevent overfitting.
    seq: (T, F)
    """
    if not training:
        return seq

    # 1. Time warping: random speed up/slow down
    if np.random.random() < 0.5:
        T = seq.shape[0]
        # Random stretch factor 0.8–1.2
        factor = np.random.uniform(0.8, 1.2)
        new_T  = int(T * factor)
        new_T  = max(4, new_T)
        idx    = np.round(np.linspace(0, T-1, new_T)).astype(int)
        seq    = resample_seq(seq[idx], T)

    # 2. Gaussian noise
    if np.random.random() < 0.5:
        noise = np.random.normal(0, 0.02, seq.shape).astype(np.float32)
        seq   = seq + noise

    # 3. Random frame dropout (simulate missing frames)
    if np.random.random() < 0.3:
        T       = seq.shape[0]
        n_drop  = np.random.randint(1, max(2, T // 8))
        drop_idx = np.random.choice(T, n_drop, replace=False)
        seq     = seq.copy()
        seq[drop_idx] = 0.0

    # 4. Horizontal flip (mirror signing)
    if np.random.random() < 0.5:
        # Since Z-axis was removed, we only have (x,y) pairs.
        # We need to selectively flip only the 'x' coordinates.

        # These are global constants from the config/feature engineering cells
        N_HAND = 21
        NEW_DIM_PER_LANDMARK = 2 # x, y
        # The spatial block size as computed in engineer_features
        FEAT_DIM_SPATIAL = (N_HAND * NEW_DIM_PER_LANDMARK) + \
                           (N_HAND * NEW_DIM_PER_LANDMARK) + \
                           (17 * NEW_DIM_PER_LANDMARK)

        # Collect indices of all x-coordinates in the feature vector
        spatial_x_indices = []

        # --- Spatial features (lh_rel, rh_rel, pose_sel) ---
        # lh_rel (N_HAND landmarks * 2 dimensions = 42 features) - x at 0, 2, ..., 40
        spatial_x_indices.extend(range(0, N_HAND * NEW_DIM_PER_LANDMARK, NEW_DIM_PER_LANDMARK))
        # rh_rel (N_HAND landmarks * 2 dimensions = 42 features) - x at 42, 44, ..., 82
        spatial_x_indices.extend(range(N_HAND * NEW_DIM_PER_LANDMARK, 2 * N_HAND * NEW_DIM_PER_LANDMARK, NEW_DIM_PER_LANDMARK))
        # pose_sel (17 landmarks * 2 dimensions = 34 features) - x at 84, 86, ..., 116
        spatial_x_indices.extend(range(2 * N_HAND * NEW_DIM_PER_LANDMARK, FEAT_DIM_SPATIAL, NEW_DIM_PER_LANDMARK))

        x_indices = list(spatial_x_indices)

        # --- Velocity features (shifted by FEAT_DIM_SPATIAL) ---
        x_indices.extend([idx + FEAT_DIM_SPATIAL for idx in spatial_x_indices])

        # --- Acceleration features (shifted by 2 * FEAT_DIM_SPATIAL) ---
        x_indices.extend([idx + 2 * FEAT_DIM_SPATIAL for idx in spatial_x_indices])

        # --- Absolute wrist positions (lh_x, rh_x) ---
        # These are at FEAT_DIM_SPATIAL*3 + 0 and FEAT_DIM_SPATIAL*3 + 2
        wrist_pos_start = 3 * FEAT_DIM_SPATIAL
        x_indices.extend([wrist_pos_start + 0, wrist_pos_start + 2])

        # --- Absolute wrist velocities (lh_vx, rh_vx) ---
        # These are at FEAT_DIM_SPATIAL*3 + 4 + 0 and FEAT_DIM_SPATIAL*3 + 4 + 2
        wrist_vel_start = 3 * FEAT_DIM_SPATIAL + 4 # 4 comes from 2x y and 2x x
        x_indices.extend([wrist_vel_start + 0, wrist_vel_start + 2])

        # Negate only the identified x-coordinates
        seq = seq.copy()
        seq[:, x_indices] *= -1.0

    return seq.astype(np.float32)
```

```python
# ============================================================
# PyTorch DATASET WITH AUGMENTATION
# ============================================================
class SignLanguageDataset(Dataset):
    def __init__(self, X, y, training=False):
        self.X = X
        self.y = y
        self.training = training

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        x_seq = self.X[idx]
        label = self.y[idx]

        if self.training:
            x_seq = augment_sequence(x_seq, training=True)

        return torch.tensor(x_seq, dtype=torch.float32), torch.tensor(label, dtype=torch.long)

def make_dataloader(X, y, training=False, batch_size=32):
    dataset = SignLanguageDataset(X, y, training=training)
    return DataLoader(dataset, batch_size=batch_size, shuffle=training, num_workers=2 if training else 0)
```

## Methodology: Model Architectures

In order to accurately represent the sequential dynamics of the dataset, I developed and assessed a collection of six unique neural network architectures. These models were chosen to capitalize on a variety of inductive biases, including global self-attention mechanisms, multi-scale temporal convolutions, and recurrent sequential processing.

All models generate unnormalized logits for `n_cls` classes by processing input sequences of the form `(Batch Size, Sequence Length, Feature Dimension)`.

### 1. Multi-Scale Convolutional LSTM with Attention (`ConvLSTM_v2`)

Local feature extraction and long-range recurrent modeling are integrated in the `ConvLSTM_v2` architecture, which is further enhanced by a custom attention mechanism.
* **Multi-Scale Feature Extraction:** The input sequence is initially processed through three parallel 1D convolutional layers with progressively increasing dilation rates (1, 2, and 4). This functions as a multi-scale temporal receptive field, enabling the network to simultaneously capture both fine-grained, immediate local changes and broader temporal trends.
* **Recurrent Processing:** The concatenated convolutional features are projected into a lower-dimensional space and subsequently input into a two-layer Bidirectional Long Short-Term Memory (Bi-LSTM) network.
* **Temporal Attention:** The network implements a softmax-normalized attention mechanism that is learnt, as opposed to conventional pooling. This layer calculates an importance weight for each timestep, generating a context vector that is a weighted sum of the hidden states of the LSTM. This ensures that the classifier concentrates solely on the most significant portions of the sequence.

### 2. Pure Self-Attention Architecture (`TransformerModel`)

In order to capture global dependencies across the sequence, this model exclusively employs multi-head self-attention, avoiding convolutions and recurrence.

* **Embeddings:** Input features are linearly projected, and learned positional embeddings are added to incorporate sequence order information—a necessity due to the fundamental permutation-invariance of self-attention operations.
* **Encoder Block:** A 4-layer standard Transformer Encoder with four attention centers per layer comprises the core. This enables the model to represent intricate, non-local relationships between any two timesteps in the sequence.
* **Aggregation:** Global average pooling is employed to consolidate the sequence of transformed embeddings into a single feature vector, which is then transmitted to the classifier.

### 3. Temporal Convolutional Network (`TCNModel`)

In contrast to RNNs, the TCN architecture is intended to capture extremely lengthy sequence dependencies through rigorously convolutional operations, thereby providing highly parallelizable training.

* **Exponential Receptive Field:** The model combines numerous residual blocks, each of which contains two 1D convolutional layers. It is crucial to note that the dilation factor doubles with each subsequent block (1, 2, 4, 8, 16, 32). This exponential growth enables the network's receptive field to encompass the entire sequence length without necessitating excessively deep architectures.
* **Strategy for Dual-Pooling:** The final output aggregates features using both global average pooling and global max pooling to ensure that no critical temporal signal is lost. A robust feature vector for classification is generated by concatenating these representations.

### 4. Modernized Recurrent Neural Network (ModernRNN)

This architecture modernizes the classic Gated Recurrent Unit (GRU) by incorporating techniques from Transformer design, despite the fact that standard RNNs frequently encounter difficulties with vanishing gradients over extended sequences.
* **Stabilized Recurrence:** Two bidirectional GRU layers are implemented in the network. In contrast to conventional GRUs, it implements Layer Normalization and residual (skip) connections following each recurrent block.
* **Advantages:** In comparison to its vanilla counterparts, these enhancements stabilize the hidden state variance across time and depth, enabling the recurrent network to train more efficiently and retain lengthier memory traces.

### 5. Hierarchical Hybrid Architecture (`HybridModel`)

Convolutions, self-attention, and recurrence are three distinct inductive biases that are integrated into a singular end-to-end framework by the `HybridModel`, an advanced, hierarchical network.
* **Stage 1 (Local Context):** A 1D convolutional layer with a kernel size of 5 is used as an initial feature extractor to reduce noise and capture local patterns. This layer is stabilized by Batch Normalization and a GELU activation.
* **Stage 2 (Global Context):** The features are transmitted through a 2-layer Transformer Encoder, which is supplemented with positional embeddings. This enables the model to recognize global, long-range dependencies throughout the sequence.
* **Stage 3 (Sequential Refinement):** Lastly, the Transformer's output is processed by a Bidirectional GRU to explicitly model the step-by-step chronological evolution of the data prior to the application of global average pooling.

### 6. DeepARTMAP: A Neuro-Symbolic Approach

DeepARTMAP is a sophisticated variant of the Adaptive Resonance Theory (ART) family of neural networks. ART models are founded on the principles of self-organization and competitive learning, in contrast to conventional deep learning models (such as CNNs or Transformers) that depend on gradient-based backpropagation. The **Stability-Plasticity Dilemma** is resolved by ART models, which is their primary advantage. They are capable of acquiring 'catastrophic forgetting' of previously acquired information (stability) while simultaneously learning new, unseen information (plasticity). A new cognitive category (node) is dynamically recruited by DeepARTMAP when it encounters a new pattern that substantially diverges from its existing knowledge, rather than destructively overwriting an old one.

The term 'Deep' in DeepARTMAP denotes its hierarchical architecture. It establishes a series of categorization layers, in which the lower levels capture highly specific, fine-grained patterns, and the higher levels generalize these into broader semantic concepts. This generates a topological map of the data that is highly explicable. Furthermore, the **vigilance parameter (rho)** governs the process of learning, determining the degree to which a new sample must resemble an existing template in order to be grouped with it. A high level of vigilance compels the network to establish numerous categories that are exceedingly specific, whereas a low level of vigilance promotes the development of generalized generalizations.

DeepARTMAP provides a captivating contrast to the PyTorch-based architectures in the context of this Sign Language Recognition project. DeepARTMAP functions as a rapid, single-pass pattern matcher on the flattened feature space, in contrast to sequence models such as LSTMs and Transformers, which explicitly map out temporal dynamics step-by-step. This network is immensely rapid to train and highly resilient to noise due to its rule-based, symbolic nature, which provides a transparent alternative to the 'black box' of traditional backpropagation networking.

### Experimental Setup and Regularization

In order to reduce overfitting in these highly parameterized models, uniform regularization strategies were implemented during the training process. The models were optimized with explicit weight decay using **AdamW**. Additionally, I implemented a warm-up phase and a cosine annealing learning rate schedule to guarantee consistent convergence. Label normalization (factor of 0.1) and excessive dropout rates (up to 50%) within the dense classification heads were implemented to penalize overconfidence in the networks.



```python
# ============================================================
# MODEL BUILDERS — PyTorch architectures & ART
# ============================================================
import numpy as np
from artlib.hierarchical import DeepARTMAP
INPUT_SHAPE = (MAX_SEQ_LEN, FEAT_DIM)

class ConvLSTM_v2(nn.Module):
    def __init__(self, n_cls, feat_dim):
        super().__init__()
        self.conv1 = nn.Conv1d(feat_dim, 128, 3, padding=1)
        self.conv2 = nn.Conv1d(feat_dim, 128, 3, padding=2, dilation=2)
        self.conv3 = nn.Conv1d(feat_dim, 128, 3, padding=4, dilation=4)
        self.proj = nn.Conv1d(384, 256, 1)

        self.lstm1 = nn.LSTM(256, 128, bidirectional=True, batch_first=True)
        self.lstm2 = nn.LSTM(256, 64, bidirectional=True, batch_first=True)
        self.attn = nn.Linear(128, 1)

        self.classifier = nn.Sequential(
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, n_cls)
        )

    def forward(self, x):
        x_t = x.transpose(1, 2) # (B, C, T)
        s1 = F.relu(self.conv1(x_t))
        s2 = F.relu(self.conv2(x_t))
        s3 = F.relu(self.conv3(x_t))
        c = torch.cat([s1, s2, s3], dim=1)
        p = F.relu(self.proj(c)).transpose(1, 2) # (B, T, 256)

        l1, _ = self.lstm1(p)
        l2, _ = self.lstm2(l1)

        attn_w = F.softmax(self.attn(l2), dim=1) # (B, T, 1)
        feat = (l2 * attn_w).sum(dim=1) # (B, 128)

        return self.classifier(feat)

class TransformerModel(nn.Module):
    def __init__(self, n_cls, feat_dim, seq_len):
        super().__init__()
        self.embed = nn.Linear(feat_dim, 128)
        self.pos_embed = nn.Parameter(torch.randn(1, seq_len, 128))
        encoder_layer = nn.TransformerEncoderLayer(d_model=128, nhead=4, dim_feedforward=512, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=4)

        self.classifier = nn.Sequential(
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, n_cls)
        )

    def forward(self, x):
        e = self.embed(x) + self.pos_embed
        t = self.transformer(e)
        pooled = t.mean(dim=1)
        return self.classifier(pooled)

class TCNModel(nn.Module):
    def __init__(self, n_cls, feat_dim):
        super().__init__()
        self.proj = nn.Conv1d(feat_dim, 128, 1)
        self.blocks = nn.ModuleList()
        for d in [1, 2, 4, 8, 16, 32]:
            self.blocks.append(nn.Sequential(
                nn.Conv1d(128, 128, 3, padding=d, dilation=d),
                nn.ReLU(),
                nn.Dropout(0.2), # Increased dropout for regularization
                nn.Conv1d(128, 128, 3, padding=d, dilation=d),
                nn.ReLU()
            ))

        self.classifier = nn.Sequential(
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(128, n_cls)
        )

    def forward(self, x):
        x_t = x.transpose(1, 2)
        out = F.relu(self.proj(x_t))
        skips = []
        for b in self.blocks:
            res = out
            out = b(out) + res
            skips.append(out)

        summed = sum(skips)
        mean_pool = summed.mean(dim=-1)
        max_pool = summed.max(dim=-1)[0]
        feat = torch.cat([mean_pool, max_pool], dim=1)
        return self.classifier(feat)

class ModernRNN(nn.Module):
    def __init__(self, n_cls, feat_dim):
        super().__init__()
        self.proj = nn.Linear(feat_dim, 256)
        self.gru1 = nn.GRU(256, 128, bidirectional=True, batch_first=True)
        self.norm1 = nn.LayerNorm(256)
        self.gru2 = nn.GRU(256, 128, bidirectional=True, batch_first=True)
        self.norm2 = nn.LayerNorm(256)

        self.classifier = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(128, n_cls)
        )

    def forward(self, x):
        x = self.proj(x)
        o1, _ = self.gru1(x)
        o1 = self.norm1(o1 + x)
        o2, _ = self.gru2(o1)
        o2 = self.norm2(o2 + o1)
        pooled = o2.mean(dim=1)
        return self.classifier(pooled)

class HybridModel(nn.Module):
    def __init__(self, n_cls, feat_dim, seq_len):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv1d(feat_dim, 256, 5, padding=2),
            nn.BatchNorm1d(256),
            nn.GELU()
        )
        self.pos_embed = nn.Parameter(torch.randn(1, seq_len, 256))

        encoder_layer = nn.TransformerEncoderLayer(d_model=256, nhead=4, dim_feedforward=512, batch_first=True, dropout=0.3)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)

        self.gru = nn.GRU(256, 128, bidirectional=True, batch_first=True)

        self.classifier = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(128, n_cls)
        )

    def forward(self, x):
        x_t = x.transpose(1, 2)
        c = self.conv(x_t).transpose(1, 2)
        c = c + self.pos_embed
        t = self.transformer(c)
        g, _ = self.gru(t)
        pooled = g.mean(dim=1)
        return self.classifier(pooled)

# ============================================================
# TRAINING HELPER
# ============================================================
def cosine_decay_with_warmup(epoch, total_epochs=80, warmup_epochs=5, base_lr=1e-3):
    if epoch < warmup_epochs:
        return base_lr * (epoch + 1) / warmup_epochs
    progress = (epoch - warmup_epochs) / (total_epochs - warmup_epochs)
    return base_lr * 0.5 * (1 + np.cos(np.pi * progress))

def train_one(model, tag, X_tr, y_tr, X_vl, y_vl):
    model.to(device)
    # Fine-Tuning: Switch to AdamW and add weight_decay to mitigate overfitting
    optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-2)
    # Fine-Tuning: Add label smoothing to prevent overconfident predictions
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

    train_loader = make_dataloader(X_tr, y_tr, training=True, batch_size=BATCH_SIZE)
    val_loader = make_dataloader(X_vl, y_vl, training=False, batch_size=BATCH_SIZE)

    best_acc = 0.0
    history = {'accuracy': [], 'val_accuracy': [], 'loss': [], 'val_loss': [], 'top5': [], 'val_top5': []}

    for epoch in range(EPOCHS):
        model.train()
        lr = cosine_decay_with_warmup(epoch, EPOCHS, 5, LR)
        for param_group in optimizer.param_groups:
            param_group['lr'] = lr

        train_loss, train_correct = 0.0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()

            train_loss += loss.item() * inputs.size(0)
            train_correct += (outputs.argmax(1) == labels).sum().item()

        model.eval()
        val_loss, val_correct, top5_correct = 0.0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                val_correct += (outputs.argmax(1) == labels).sum().item()

                _, top5_preds = outputs.topk(5, 1, True, True)
                top5_correct += torch.sum(top5_preds == labels.unsqueeze(1)).item()

        acc = train_correct / len(X_tr)
        val_acc = val_correct / len(X_vl)
        val_top5 = top5_correct / len(X_vl)

        history['accuracy'].append(acc)
        history['val_accuracy'].append(val_acc)
        history['loss'].append(train_loss / len(X_tr))
        history['val_loss'].append(val_loss / len(X_vl))
        history['val_top5'].append(val_top5)

        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), str(CKPT_DIR / f'{tag}_best.pt'))

        if (epoch+1) % 10 == 0:
            print(f"Epoch {epoch+1:03d} | Train Acc: {acc:.4f} | Val Acc: {val_acc:.4f} | Val Top5: {val_top5:.4f}")

    model.load_state_dict(torch.load(str(CKPT_DIR / f'{tag}_best.pt')))
    print(f"  {tag} -> best_val_acc={best_acc:.4f}")

    # Generate predictions for ensembling
    model.eval()
    preds = []
    with torch.no_grad():
        for inputs, _ in val_loader:
            outputs = F.softmax(model(inputs.to(device)), dim=1)
            preds.append(outputs.cpu().numpy())
    preds = np.concatenate(preds, axis=0)

    class HistWrapper:
        def __init__(self, d):
            self.history = d
    return HistWrapper(history), float(best_acc), float(history['val_top5'][-1]), preds

def plot_hist(h, tag):
    fig, axes = plt.subplots(1, 3, figsize=(16, 4))
    for ax, k, t in zip(axes,
                         ['accuracy', 'loss', 'val_top5'],
                         ['Accuracy', 'Loss', 'Top-5']):
        if k in h.history:
            ax.plot(h.history[k], label='Train')
        if f'val_{k}' in h.history and k != 'val_top5':
            ax.plot(h.history[f'val_{k}'], label='Val')
        ax.set_title(f'{tag} – {t}')
        ax.set_xlabel('Epoch'); ax.legend(); ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(CKPT_DIR / f'{tag}_history.png', dpi=120)
    plt.close()

```

```python
# ============================================================
# [6/7] TRAIN 5 MODELS + DeepARTMAP
# ============================================================
print("\n[6/7] Training...")

print("\n━━━ Conv-LSTM v2 (with attention) ━━━")
m1 = ConvLSTM_v2(n_cls, FEAT_DIM)
h1, a1, t1, p1 = train_one(m1, 'conv_lstm_v2', X_tr, y_tr, X_vl, y_vl)
plot_hist(h1, 'ConvLSTM_v2')

print("\n━━━ Transformer ━━━")
m2 = TransformerModel(n_cls, FEAT_DIM, MAX_SEQ_LEN)
h2, a2, t2, p2 = train_one(m2, 'transformer', X_tr, y_tr, X_vl, y_vl)
plot_hist(h2, 'Transformer')

print("\n━━━ TCN ━━━")
m3 = TCNModel(n_cls, FEAT_DIM)
h3, a3, t3, p3 = train_one(m3, 'tcn', X_tr, y_tr, X_vl, y_vl)
plot_hist(h3, 'TCN')

print("\n━━━ Modern RNN (GRU) ━━━")
m4 = ModernRNN(n_cls, FEAT_DIM)
h4, a4, t4, p4 = train_one(m4, 'modern_rnn', X_tr, y_tr, X_vl, y_vl)
plot_hist(h4, 'ModernRNN')

print("\n━━━ Hybrid Model (Conv+Trans+GRU) ━━━")
m5 = HybridModel(n_cls, FEAT_DIM, MAX_SEQ_LEN)
h5, a5, t5, p5 = train_one(m5, 'hybrid_model', X_tr, y_tr, X_vl, y_vl)
plot_hist(h5, 'HybridModel')

```

```python
print("\n━━━ DeepARTMAP ━━━")
# DeepART typically requires flattened inputs and competitive training.
# We flatten the sequence for the ART model.
X_tr_flat = X_tr.reshape(X_tr.shape[0], -1)
X_vl_flat = X_vl.reshape(X_vl.shape[0], -1)

# Normalize data to [0, 1] for FuzzyART
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_tr_flat = scaler.fit_transform(X_tr_flat)
X_vl_flat = scaler.transform(X_vl_flat)

# Ensure strictly [0, 1] due to floating point inaccuracies
X_tr_flat = np.clip(X_tr_flat, 0, 1)
X_vl_flat = np.clip(X_vl_flat, 0, 1)

# Apply complement coding
X_tr_flat = np.concatenate([X_tr_flat, 1.0 - X_tr_flat], axis=1)
X_vl_flat = np.concatenate([X_vl_flat, 1.0 - X_vl_flat], axis=1)

# Cast to float64 to avoid Numba type unification errors in artlib
X_tr_flat = X_tr_flat.astype(np.float64)
X_vl_flat = X_vl_flat.astype(np.float64)

# Initialize DeepARTMAP from artlib, passing a list of modules.
try:
    from artlib.hierarchical.DeepARTMAP import DeepARTMAP
    from artlib.elementary.FuzzyART import FuzzyART

    base_module = FuzzyART(rho=0.85, alpha=0.001, beta=1.0)
    m6 = DeepARTMAP(modules=[base_module])
except Exception as e:
    print(f"Initialization failed: {e}")
    m6 = None

# Training (assuming fit/predict scikit-learn like API based on standard ART implementations)
try:
    if m6 is not None:
        # DeepARTMAP expects a list of inputs, one for each base module
        m6.fit([X_tr_flat], y_tr)
        art_preds = m6.predict_proba([X_vl_flat]) if hasattr(m6, 'predict_proba') else np.eye(n_cls)[m6.predict([X_vl_flat])]
        a6 = float(np.mean(np.argmax(art_preds, axis=1) == y_vl))
        print(f"  DeepARTMAP -> best_val_acc={a6:.4f}")
    else:
        raise ValueError("m6 was not initialized successfully.")
except Exception as e:
    print(f"  DeepARTMAP Training encountered an exception: {e}")
    a6 = 0.0
    art_preds = np.zeros((len(X_vl), n_cls))
```

```python
# ============================================================
# [7/7] ENSEMBLE + RESULTS
# ============================================================
print("\n[7/7] Ensemble & results...")

# Calculate top-5 for DeepARTMAP if not available
if 'art_preds' in locals():
    t6 = float(np.mean([y_vl[i] in np.argsort(art_preds[i])[-5:] for i in range(len(y_vl))]))
else:
    a6, t6, art_preds = 0.0, 0.0, np.zeros_like(p1)

w = np.array([a1, a2, a3, a4, a5, a6], dtype=np.float64)
w /= w.sum()
print(f"  Weights: ConvLSTM={w[0]:.3f}  Transformer={w[1]:.3f}  TCN={w[2]:.3f}  ModernRNN={w[3]:.3f}  Hybrid={w[4]:.3f}  DeepARTMAP={w[5]:.3f}")

p_ens  = w[0]*p1 + w[1]*p2 + w[2]*p3 + w[3]*p4 + w[4]*p5 + w[5]*art_preds
y_true = y_vl
y_pred = np.argmax(p_ens, axis=1)

ens_acc  = float(np.mean(y_pred == y_true))
ens_top5 = float(np.mean(
    [y_true[i] in np.argsort(p_ens[i])[-5:] for i in range(len(y_true))]
))

print(f"\n  Ensemble Acc  : {ens_acc:.4f}")
print(f"  Ensemble Top-5: {ens_top5:.4f}")

results = pd.DataFrame({
    'Model': ['ConvLSTM_v2', 'Transformer', 'TCN', 'ModernRNN', 'Hybrid', 'DeepARTMAP', 'Ensemble'],
    'Acc':   [a1, a2, a3, a4, a5, a6, ens_acc],
    'Top5':  [t1, t2, t3, t4, t5, t6, ens_top5],
})
print("\n" + results.to_string(index=False))

# ── Final plots ──────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x_ = np.arange(len(results)); bw = 0.35
axes[0].bar(x_ - bw/2, results['Acc'],  bw, label='Accuracy',  alpha=0.85)
axes[0].bar(x_ + bw/2, results['Top5'], bw, label='Top-5 Acc', alpha=0.85)
axes[0].set_xticks(x_); axes[0].set_xticklabels(results['Model'], rotation=10)
axes[0].set_ylim(0, 1.1); axes[0].set_ylabel('Score')
axes[0].set_title('Model Comparison'); axes[0].legend()
axes[0].grid(True, alpha=0.3, axis='y')

per_cls = [float(np.mean(y_pred[y_true==i]==i)) if (y_true==i).sum()>0 else 0.0
           for i in range(n_cls)]
axes[1].bar(range(n_cls), sorted(per_cls, reverse=True), alpha=0.75, width=1.0)
axes[1].axhline(ens_acc, color='red', ls='--', lw=2, label=f'Mean {ens_acc:.3f}')
axes[1].set_xlabel('Class (sorted)'); axes[1].set_ylabel('Accuracy')
axes[1].set_title('Per-Class Accuracy'); axes[1].legend()
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig(CKPT_DIR / 'final_results.png', dpi=120)
plt.close()

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(cm, cmap='Blues', ax=ax, xticklabels=False, yticklabels=False)
ax.set_xlabel('Predicted'); ax.set_ylabel('True')
ax.set_title(f'Confusion Matrix ({n_cls} classes)  Acc={ens_acc:.3f}')
plt.tight_layout()
plt.savefig(CKPT_DIR / 'confusion_matrix.png', dpi=120)
plt.close()

# ── Sample predictions ────────────────────────────────────────
print("\n" + "=" * 65)
print("SAMPLE PREDICTIONS")
print("=" * 65)
for i in np.random.choice(len(y_true), min(15, len(y_true)), replace=False):
    tl = le.inverse_transform([y_true[i]])[0]
    pl = le.inverse_transform([y_pred[i]])[0]
    c  = float(p_ens[i, y_pred[i]])
    top5_labels = le.inverse_transform(np.argsort(p_ens[i])[-5:][::-1])
    print(f"  {'✓' if tl==pl else '✗'}  True: {tl:<20s}  "
          f"Pred: {pl:<20s}  Conf: {c:.3f}")

# ── Save all ───────────────────────────────────────────────────
np.save(CKPT_DIR / 'ensemble_preds.npy', p_ens)
np.save(CKPT_DIR / 'val_labels.npy', y_vl)

with open(CKPT_DIR / 'results.json', 'w') as f:
    json.dump({
        'n_classes': int(n_cls), 'feat_dim': FEAT_DIM,
        'n_train': int(len(X_tr)), 'n_val': int(len(X_vl)),
        'conv_lstm_acc': a1, 'transformer_acc': a2, 'tcn_acc': a3,
        'modern_rnn_acc': a4, 'hybrid_acc': a5, 'deepartmap_acc': a6,
        'ensemble_acc': ens_acc, 'ensemble_top5': ens_top5,
        'weights': w.tolist(), 'label_classes': le.classes_.tolist(),
    }, f, indent=2)

print(f"\n{'='*65}")
print(f"COMPLETE — saved to {CKPT_DIR}")
print(f"{'='*65}")
```

## Conclusion

As anticipated, the recurrent architectures—particularly the ModernRNN (GRU)—demonstrated strong performance in modeling the sequential nature of the sign language data, achieving the highest individual validation accuracy (~78%). The ensemble model ultimately reached over 80% accuracy, confirming that capturing complex temporal dynamics through a combination of approaches yields the best results.

On the other hand, the neuro-symbolic DeepARTMAP model requires significantly more training time to properly establish its hierarchical cognitive categories and process this high-dimensional dataset. Given the current constraints and deadlines, its training could not be fully completed. Future work could explore optimized or parallelized implementations of ART networks to accelerate their convergence on large-scale, continuous gesture recognition tasks.
